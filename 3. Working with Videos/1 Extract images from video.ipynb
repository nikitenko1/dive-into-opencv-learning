<p style="font-family:ComicSansMS; font-size: 30px; color: magenta"> Extract images from video in Python</p>

<p style="font-family:ComicSansMS; font-size: 20px; color: orange"> Function Used :</p>

* <p style="font-family:ComicSansMS; font-size: 20px; color: orange">VideoCapture(File_path) : Read the video(.mp4 format)</<p>
* <p style="font-family:ComicSansMS; font-size: 20px; color: orange">read() : Read data depending upon the type of object that calls</<p> 
* <p style="font-family:ComicSansMS; font-size: 20px; color: orange">imwrite(filename, img[, params]) : Saves an image to a specified file.</<p>

> Below is the implementation:

```python
# Importing all necessary libraries
import cv2
import os
```

```python
# Read the video from specified path
cap = cv2.VideoCapture('../data/pexels-takeoff.mp4')

try:
    
    # creating a folder named data
    if not os.path.exists('data'):
        os.makedirs('data')

# if not created then raise error
except OSError:
    print ('Error: Creating directory of data')

# frame
currentframe = 0

while(True):
    
    # reading from frame
    ret,frame = cap.read()

    if ret:
        # if video is still left continue creating images
        name = './data/frame' + str(currentframe) + '.jpg'
        print ('Creating...' + name)

        # writing the extracted images
        cv2.imwrite(name, frame)

        # increasing counter so that it will
        # show how many frames are created
        currentframe += 1
    else:
        break

# Release all space and windows once done
cap.release()
cv2.destroyAllWindows()
```
